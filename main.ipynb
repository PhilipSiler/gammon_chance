```python
seed_xgid = 'XGID=-O----------------aaaaaai-:0:0:1:00:0:0:3:0:10'

#for i, c in enumerate(seed_xgid):
#    print(str(i) + " " + c)
```

```python
def increase_top_player_pip(xgid):
    xgid = list(xgid)
    index_of_blot = -1
    if 'a' in xgid[13:24]: #if there is a blot here that we can move back
        index_of_blot = xgid[13:24].index('a') + 13

    if index_of_blot == 13: #move blot to checker stack on 7 point
        if xgid[12] == '-': #edge case, first increment of 7 point
            char_to_increment = 'a'
        else: #increment up the char that's already on the 7 point
            char_to_increment = xgid[12]
            char_to_increment = chr(ord(char_to_increment) + 1)
        xgid[12] = char_to_increment
        xgid[13] = '-' #remove checker on 

    elif index_of_blot == -1: #move blot from checker stack on 24 point to 17
        char_to_decrement = xgid[29]
        char_to_decrement = chr(ord(char_to_decrement) - 1)
        xgid[29] = char_to_decrement
        xgid[23] = 'a'

    else: #general case, move blot one space back
        xgid[index_of_blot] = '-'
        xgid[index_of_blot - 1] = 'a'

    return "".join(xgid)
```

```python
def decrease_bottom_player_rolls(xgid):
    xgid = list(xgid)
    xgid[6] = chr(ord(xgid[6]) - 2)
    return "".join(xgid)
```

```python
cur_xgid = seed_xgid
print(cur_xgid)
for _ in range(100):
    cur_xgid = increase_top_player_pip(cur_xgid)
```

```python
cur_xgid = seed_xgid
print(cur_xgid)
for _ in range(7):
    cur_xgid = decrease_bottom_player_rolls(cur_xgid)
```
